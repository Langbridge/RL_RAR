```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = nx.grid_graph(dim=(5,4))
G = nx.convert_node_labels_to_integers(G)
G = nx.DiGraph(G)

pos = nx.spring_layout(G, iterations=100)
```

```python
S = 13
E = 6

a = np.zeros((20,))
idxs = [y for (x, y) in G.edges(S)]
a[idxs] = 1
a
```

```python
options = {
    "font_size": 14,
    "node_size": 3000,
    "edgecolors": "black",
    "linewidths": 5,
    "width": 5,
}

plt.figure(figsize=(10,10))

nx.draw_networkx(G, pos=pos, node_color='white', **options)
nx.draw_networkx(G, nodelist=[S], pos=pos, node_color='blue', alpha=0.4, label='Start', **options)
nx.draw_networkx(G, nodelist=[E], pos=pos, node_color='green', alpha=0.4, label='End', **options)
plt.legend(labelspacing=5)
```
